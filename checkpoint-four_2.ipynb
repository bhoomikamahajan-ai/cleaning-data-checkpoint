```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv(r"C:\Users\User\group-challenge\projects\cleaning-data-checkpoint\output.csv")
```

```python
df.drop(columns = ["StartDate", "Duration", "Location"] , inplace = True)
```

```python
df = df.drop(df[df["JobDescription"].str.lower() == "unknown"].index)
```

```python
df.shape
```

```python
df.drop(columns=["JobRequirment", "JobDescription"], inplace= True)
```

```python
df.shape
```

```python
skills_exploded = df.explode("Matched_Skills")  # This converts each list of skills into individual rows.

skills_exploded = skills_exploded[
    skills_exploded["Matched_Skills"].notna() &
    (skills_exploded["Matched_Skills"] != "")
]


skill_counts = (
    skills_exploded["Matched_Skills"]
    .value_counts()
    .reset_index()
)

skill_counts.columns = ["Skill", "Count"]
```

```python
skill_counts = skill_counts.sort_values("Count", ascending=False)
```

```python
skill_counts.to_csv("skill_counts_tableau.csv", index=False)
```

```python
# Remove all non-alphanumeric characters (except space), replace with a space, and convert to lowercase
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str) \
                     .str.replace(r'[^a-zA-Z0-9\s]', ' ', regex=True) \
                     .str.lower() \
                     .str.strip()
                     #.str.replace(r'\s+', ' ', regex=True) # Optional
```

```python
df.to_csv("final_tableau.csv",  index=False)
```
