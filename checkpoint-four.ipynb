```python
df["RequiredQual"] = df["RequiredQual"].fillna("").str.lower()
```

```python
# This is a an upadte from the check point three. Here instead of just checking the skills and saving them to the a new column, I am created an new column for each skill and using an flag to indicate if the skill is present in for that role. 
# I had to alter the approach because the vizzs are more clear like this. 
```

```python
skills_list = ["python", "java", "c++", "javascript", "php", "core java","html", "css", "web", "apis", "rest services","data structures", "sql", "nosql", "databases",
                "git", "github","software development lifecycle", "testing", "agile", "scrum", "waterfall","financial reporting", "gaap", "ledger", "tax", "bookkeeping", 
                "budgeting", "risk management","quickbooks", "sap", "oracle","legal research", "contract drafting", "litigation procedures", "case management systems"]
for skill in skills_list:
    col_name = skill.replace(" ", "_").replace("+", "plus")
    df[col_name] = df["RequiredQual"].str.contains(skill, regex=False).astype(int)
```

```python
skill_counts = (
    df[[skill.replace(" ", "_").replace("+", "plus") for skill in skills_list]]
    .sum()
    .reset_index()
)

skill_counts.columns = ["Skill", "Count"]
```

In the below code, removing all kind of delimiters from the file, because Tableau treated them as a field separator and created multiple columns and rows. 

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str) \
                     .str.replace(r'[^a-zA-Z0-9\s]', ' ', regex=True) \
                     .str.lower() \
                     .str.strip()
                     
```

```python
df.to_csv("jobs_with_skill_flags.csv", index=False)
```
